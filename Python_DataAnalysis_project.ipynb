### DATA ANALYSIS

#### 1.How many are there in each Team and the percentage splitting with respect to the total employees.

```python
#number of employees in each Team

counts = df1["Team"].value_counts()
print("No of employees in each team:\n",counts)
```

```python
#Visualization

plt.figure(figsize=(60,30),dpi=100)
x=counts.index
y=counts.values
plt.bar(x, y, color =['red','blue'],width=.2)
plt.xticks(fontsize=35)
plt.xticks(rotation=45)
plt.yticks(fontsize=35)
plt.title("Number of Employees in each team",size=50)
plt.xlabel("Team",size=50)
plt.ylabel("Number of Employees",size=50)
plt.show()
```

```python
#percentage splitting of employees with respect to the total employees.

percentage = df1["Team"].value_counts(normalize=True).mul(100).round(1)
print("The percentage splitting of employees with respect to the total employees:\n",percentage)
```

```python
#visualization

plt.figure(figsize=(60,30),dpi=100)
x=percentage.index
y=percentage.values
plt.bar(x, y, color =['red','blue'],width=.2)
plt.xticks(fontsize=35)
plt.xticks(rotation=45)
plt.yticks(fontsize=35)
plt.title("Percentage split of Employees by team",size=50)
plt.xlabel("Team",size=50)
plt.ylabel("Percentage",size=50)
plt.show()
```

#### 2.Segregate the employees w.r.t different positions.

#### 3.Find from which age group most of the employees belong to.

```python
#Visualization

plt.figure(figsize=(5,5),dpi=100)
x=age.index
y=age.values
plt.bar(x, y, color =['red','blue'],width=.4)
plt.title("Age Vs Employees")
plt.xlabel("Age")
plt.ylabel("Number of Employees")
plt.show()
```

#### 4.Find out under which team and position, spending in terms of salary is high

```python
salary_sum=df1.groupby(['Team','Position'])['Salary'].sum()
print(salary_sum)
```

```python
#team and position spending high salary

heighest_salary = salary_sum.idxmax()
print("highest salary is:",heighest_salary)
print("under {} team and {} position, spending in terms of salary is high".format (heighest_salary[0],heighest_salary[1]))
```

#### 5.Find if there is any correlation between age and salary , represent it visually.

```python
correlation=df1['Age'].corr(df1["Salary"])
print("The correlation between age and salary is",correlation)

#Visualization

sns.scatterplot(data=df1,x='Age',y='Salary')
plt.title("correlation between age and salary")
plt.show()
```
